> To answer my first question, I am interested in exploring the 'Segment_Category' section of the TOP 250 restaurants:





```
display(df_top250[['Restaurant','Segment_Category']])
```



> I realized that the form of categorization in the original dataframe is not very informative, as it mixed the information about the type of restaurants and the type of food it served. So I decide to split it into two other columns to better analyze the data:



```
# To better Answer Question 1:

# Add a 'Grouped_Category' column to the df_top250 dataframe to further
# categorize different types of restaurants
df_top250['Grouped_Category'] = ""
for row in range(250):
  if df_top250.iloc[row]['Segment_Category'].startswith('Quick Service'):
    df_top250.at[row,'Grouped_Category'] = 'Quick Service'
  elif df_top250.iloc[row]['Segment_Category'].startswith('Casual Dining'):
    df_top250.at[row,'Grouped_Category'] = 'Casual Dining'
  elif df_top250.iloc[row]['Segment_Category'].startswith('Family'):
    df_top250.at[row,'Grouped_Category'] = 'Family Style'
  elif df_top250.iloc[row]['Segment_Category'].startswith('Fast Casual'):
    df_top250.at[row,'Grouped_Category'] = 'Fast Casual'
  else:
    df_top250.at[row,'Grouped_Category'] = 'Others'

# Add a 'Served_Food' column to the df_top250 dataframe to learn about the
# main type of food that is served in each restaurant
df_top250['Served_Food'] = ""
for row in range(250):
  if "Varied" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Varied'
  elif "Burger" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Burger'
  elif "Cafe" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Cafe Style'
  elif "Chicken" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Chicken'
  elif "Mexican" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Mexican Style'
  elif "Sandwich" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Sandwich'
  elif "Pizza" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Pizza'
  elif "Asian" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Asian Style'
  elif "Seafood" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Seafood'
  elif "Steak" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Steak'
  elif "BBQ" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'BBQ'
  elif "Desserts" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Desserts'
  elif "Family" in df_top250.iloc[row]['Segment_Category']:
    df_top250.at[row,'Served_Food'] = 'Family Style'
  else:
    df_top250.at[row,'Served_Food'] = 'Others'

display(df_top250[['Restaurant', 'Grouped_Category', 'Served_Food']])

  
```

**Phase 2: Data Visualization and Question Investigation**

Question 1: What type of restaurants were getting more popular in 2020?

```
# group number of restaurants by categories
df_category_count = df_top250.groupby(by='Grouped_Category').count()
df_category_count.sort_values(by='Restaurant', inplace=True)
df_category_count.drop(index='Others', inplace=True)
display(df_category_count[['Restaurant']])
```

```
# visualization: restaurants by categories
df_category_count['Restaurant'].plot(kind='bar', figsize=(15,9),
                                     fontsize = 12, color= 'brown')
plt.xticks(rotation=0)
plt.xlabel('Restaurant Category', color= 'brown', fontsize = 15)
plt.ylabel('Restaurant Counts', color= 'brown', fontsize = 15)
plt.title('The Most Popular Restaurants by Category', 
          color= 'brown', fontsize = 20)
plt.show()
```



> Subquestion 1-1: What is the proportion of quick service restaurants among the Top 30 restaurants?



```
# Find out the proportion of quick service restaurants among the Top 30 restaurants
df_category_proportion = df_top250
df_category_proportion['isQuickService'] = df_category_proportion['Grouped_Category'] == 'Quick Service'
df_category_proportion = df_category_proportion.loc[0:29]
display(df_category_proportion[['Rank', 'Restaurant', 'isQuickService']])
```

```
# Find out the proportion of quick service restaurants among the Top 30 restaurants
df_category_proportion = df_category_proportion.groupby(by='isQuickService').count()
display(df_category_proportion[['Restaurant']])
```

```
# Visualization: the proportion of quick service restaurants among the Top 30 restaurants
proportion_label = ['Other Restaurants', 'Quick Service Restaurants']
df_category_proportion['Restaurant'].plot(kind='pie',
                                          labels = proportion_label,
                                          autopct='%.f%%',
                                          figsize=(8,8),
                                          fontsize = 10)
plt.ylabel('')
plt.title('Proportion of Quick Service Restaurants among the Top 30',
          fontsize=15)
plt.axis('equal')
plt.show()
```



> Subquestion 1-3: What are the average and total sales of different types of restaurants?



```
# Calculate the average and total sales of different types of restaurants
df_category_sales = df_top250[['Restaurant','Sales','Grouped_Category']]

df_category_avg = df_category_sales.groupby(by='Grouped_Category').mean()
df_category_sum = df_category_sales.groupby(by='Grouped_Category').sum()

df_category_avg.rename(columns= {'Sales': 'Average Sales (in million dollars)'}, inplace=True)
df_category_sum.rename(columns= {'Sales': 'Total Sales (in million dollars)'}, inplace=True)

df_category_avg.drop(index='Others', inplace=True)
df_category_sum.drop(index='Others', inplace=True)

display(df_category_avg)
display(df_category_sum)
```

```
# visualization: average sales of restaurants by category
df_category_avg['Average Sales (in million dollars)'].plot(kind='bar', 
                                                           figsize=(15,9),
                                                           fontsize = 12, 
                                                           color= 'purple')
plt.xticks(rotation=0)
plt.xlabel('Restaurant Category', color= 'purple', fontsize = 15)
plt.ylabel('Average Sales (in million dollars)', color= 'purple', fontsize = 15)
plt.title('Average Sales of the Top 250 Restaurants by Category', 
          color= 'purple', fontsize = 20)
plt.show()
```

```
# visualization: total sales of restaurants by category
df_category_sum['Total Sales (in million dollars)'].plot(kind='bar', 
                                                           figsize=(15,9),
                                                           fontsize = 12, 
                                                           color= 'olive')
plt.xticks(rotation=0)
plt.xlabel('Restaurant Category', color= 'olive', fontsize = 15)
plt.ylabel('Total Sales (in million dollars)', color= 'olive', fontsize = 15)
plt.title('Total Sales of the Top 250 Restaurants by Category', 
          color= 'olive', fontsize = 20)
plt.show()
```



> Subquestion 1-4: What types of food are served by the most popular resaurants?



```
# group number of restaurants by type of food
df_category_food = df_top250.groupby(by='Served_Food').count()
df_category_food.sort_values(by='Restaurant', inplace=True)
df_category_food.drop(index='Others', inplace=True)
display(df_category_food[['Restaurant']])
```

```
# visualization: restaurants by type of food
df_category_food['Restaurant'].plot(kind='bar', figsize=(15,9),
                                     fontsize = 12, color= 'darkblue')
plt.xticks(rotation=45)
plt.xlabel('Type of Food Served', color= 'darkblue', fontsize = 15)
plt.ylabel('Restaurant Counts', color= 'darkblue', fontsize = 15)
plt.title('The Most Popular Restaurants by Type of Food Served', 
          color= 'darkblue', fontsize = 20)
plt.show()
```

Question 2: What was the spread of the top-ranked restaurants, in terms of locations?

```
# by state
df_ind100_state = df_ind100.groupby(by='State').count()
df_ind100_state = df_ind100_state[['Restaurant']]
df_ind100_state.sort_values(by='Restaurant' ,inplace=True)
display(df_ind100_state)
```

```
# by Top 5 cities
df_ind100_city = df_ind100.groupby(by='City').count()
df_ind100_city = df_ind100_city[['Restaurant']]
df_ind100_city.sort_values(by='Restaurant', ascending=False, inplace=True)
df_ind100_city.reset_index(inplace=True)
df_ind100_city = df_ind100_city.loc[0:4]
display(df_ind100_city)
```

```
# visualization: restaurants by states
df_ind100_state['Restaurant'].plot(kind='bar', figsize=(15,9),
                                     fontsize = 12, color= 'seagreen')
plt.xticks(rotation=90)
plt.yticks(np.arange(0, 24, 2))

plt.xlabel('State', color= 'seagreen', fontsize = 15)
plt.ylabel('Restaurant Counts', color= 'seagreen', fontsize = 15)
plt.title('The Most Popular Restaurants by States', 
          color= 'seagreen', fontsize = 20)
plt.show()
```

Question 3: How did the COVID-19 negatively (or potentially positively) affect the restaurant industry in the United States?

```
# visualization: YOY_Sales Statistics
top250_YOY_Sales = df_top250[['YOY_Sales']]
top250_YOY_Sales['YOY_Sales'] = top250_YOY_Sales['YOY_Sales'].str.rstrip('%').astype('float') / 100.0
top250_YOY_Sales['YOY_Sales'].plot(kind='box', figsize=(15,3), vert=False)
plt.xticks(np.arange(-0.2, 0.4, step=0.1), ['-20%','-10%','0%','10%','20%','30%','40%'])
plt.xlabel('Percentage', color= 'steelblue', fontsize = 15)
plt.title('Top 250 Restaurants Year on Year Sales Increase', 
          color= 'steelblue', fontsize = 20)
plt.show()
```

```
# visualization: YOY_Units Statistics
top250_YOY_Units = df_top250[['YOY_Units']]
top250_YOY_Units['YOY_Units'] = top250_YOY_Units['YOY_Units'].str.rstrip('%').astype('float') / 100.0
top250_YOY_Units['YOY_Units'].plot(kind='box', figsize=(15,3), vert=False)
plt.xticks(np.arange(-0.2, 0.4, step=0.1), ['-20%','-10%','0%','10%','20%','30%','40%'])
plt.xlabel('Percentage', color= 'navy', fontsize = 15)
plt.title('Top 250 Restaurants Year on Year Units Increase', 
          color= 'navy', fontsize = 20)
plt.show()
```

Question 4: Among the restaurants with most future potentials, is franchising a popular practice?

```
# group the restaurants by Franchising
df_futr50_franchising = df_futr50.groupby(by='Franchising').count()
df_futr50_franchising = df_futr50_franchising[['Restaurant']]
display(df_futr50_franchising)
```

An Additional Question upon Data Exploration:
*   I'm curious to see if there is a relationship between the restaurants' sales volume and the number of premises they have...







```
df_relationship = df_top250[['Sales','Units']]
plt.figure(figsize=(15,9))
plt.scatter(df_relationship['Sales'],df_relationship['Units'])

z = np.polyfit(np.log(df_relationship['Sales']), np.log(df_relationship['Units']), 1)
p = np.poly1d(z)
plt.plot(df_relationship['Sales'], p(df_relationship['Sales']), "r--")

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Sales (in million dollars, log-scaled)', color= 'navy', fontsize = 15)
plt.ylabel('Units (log-scaled)', color= 'navy', fontsize = 15)
plt.title('Relationship between Restaurants Sales and Units', 
          color= 'navy', fontsize = 20)
plt.show()
```
